```python
#Violin Plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#z = sns.violinplot(data = movies, x = 'genre', y = 'criticsRating')
z = sns.violinplot(data = movies[movies.genre=='Drama'], x = 'release', y = 'criticsRating')
```

```python
#z1 = sns.boxplot(data = movies, x = 'genre', y = 'criticsRating')
z1 = sns.boxplot(data = movies[movies.genre=='Drama'], x = 'release', y = 'criticsRating')
```

```python
#Create a facet grid
```

```python
g = sns.FacetGrid(movies, row = 'genre', col = 'release', hue = 'genre')
g = g.map(plt.scatter, 'criticsRating', 'audienceRating')
#plt.scatter(movies.criticsRating, movies.audienceRating)
```

```python
#can populate with any type of chart
```

```python
g = sns.FacetGrid(movies, row = 'genre', col = 'release', hue = 'genre')
g = g.map(plt.hist, 'budgetInMillions')
```

```python
g = sns.FacetGrid(movies, row = 'genre', col = 'release', hue = 'genre',)
kws = dict(s=50,linewidth=0.5,edgecolor='black')
g = g.map(plt.scatter, 'criticsRating', 'audienceRating', **kws)
```

```python
g = sns.FacetGrid(movies, row = 'genre', col = 'release', hue = 'genre',)
kws = dict(s=50,linewidth=0.5,edgecolor='black')
g = g.map(plt.scatter, 'criticsRating', 'audienceRating', **kws)
g.set(xlim=(0,100),ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='grey',ls='--')
g.add_legend()
```
